```python
import math
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('diamonds.csv')
df.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(list(df['fancy_color_intensity']))
df['fancy_color_intensity_pre'] = df['fancy_color_intensity'].apply(lambda x: le.transform([x])[0])
df[['fancy_color_intensity_pre', 'fancy_color_intensity']]
```
