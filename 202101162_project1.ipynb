# Preprocessing

```python
from sklearn import preprocessing
le_cut = preprocessing.LabelEncoder()
le_cut.fit(list(df['cut']))
df['cut_pre'] = df['cut'].apply(lambda x: le_cut.transform([x])[0])
df[['cut', 'cut_pre']]
```

```python
df.columns
```

```python
le_color = preprocessing.LabelEncoder()
le_color.fit(list(df['color']))
df['color_pre'] = df['color'].apply(lambda x: le_color.transform([x])[0])
df[['color', 'color_pre']]
```

```python
le_clarity = preprocessing.LabelEncoder()
le_clarity.fit(list(df['clarity']))
df['clarity_pre'] = df['clarity'].apply(lambda x: le_clarity.transform([x])[0])
df[['clarity', 'clarity_pre']]
```

```python
le_cut_quality = preprocessing.LabelEncoder()
le_cut_quality.fit(list(df['cut_quality']))
df['cut_quality_pre'] = df['cut_quality'].apply(lambda x: le_cut_quality.transform([x])[0])
df[['cut_quality', 'cut_quality_pre']]
```

```python
le_lab = preprocessing.LabelEncoder()
le_lab.fit(list(df['lab']))
df['lab_pre'] = df['lab'].apply(lambda x: le_lab.transform([x])[0])
df[['lab', 'lab_pre']]
```

```python
le_symmetry = preprocessing.LabelEncoder()
le_symmetry.fit(list(df['symmetry']))
df['symmetry_pre'] = df['symmetry'].apply(lambda x: le_symmetry.transform([x])[0])
df[['symmetry', 'symmetry_pre']]
```
