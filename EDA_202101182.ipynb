```python
# list of categorical variables
categorical_columns = [feature for feature in df.columns if df[feature].dtypes =='O']
```

```python
# list of categorical variables
numerical_columns=[feature for feature in df.columns if df[feature].dtypes !='O']
numerical_columns.pop(-1)
features=numerical_columns+categorical_columns
```

```python
#our target
target=['total_sales_price']
print('numerical_features:{}'.format(numerical_columns))
print('---------- ')
print('categorical_features:{}'.format(categorical_columns))
print('-----------')
print('target:{}'.format(target))
```

<h1>A few points to notice in these pair plots<h1>

There are some features with datapoint that are far from the rest of the dataset which will affect the outcome of our regression model.

```python
#relation between total sale price and the others categorical variables
for cat in categorical_columns:
    sns.barplot(df,x='total_sales_price',y=cat)
    plt.xticks(rotation=75)
    plt.title(cat)
    print()

    plt.show()
```

```python
#relation between total sale price and the others numerical variables
for cat in numerical_columns:
    sns.relplot(df,x='total_sales_price',y=cat, hue= 'cut_quality' )
    plt.xticks(rotation=75)
    plt.title(cat)

    plt.show()
    print()
```

```python
# Distrubitions
for cat in numerical_columns:
    sns.kdeplot(df,x=cat)
    plt.xticks(rotation=75)
    plt.title('Distrubition of {}'.format(cat))
    plt.show()
    print()
sns.kdeplot(df,x='total_sales_price',gridsize=100)
plt.title('Distrubition of total_sales_price')
plt.show()
```
